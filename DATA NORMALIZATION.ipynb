```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
```

# DATA NORMALIZATION TECHNIQUES USING REGULAR PYTHON

### importing dataset:

```python
boston = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
```

```python
boston.info()
```

```python
boston.describe()
```

```python
boston.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston['crim'], bins=30)
plt.show()
```

## Normalize data using max-min technique:

```python
boston_1 = boston.apply(lambda x:(x-x.min(axis=0))/(x.max(axis=0)-x.min(axis=0)))
```

```python
boston_1.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston_1['crim'], bins=30)
plt.show()
```

## Normalize data using mean normalization technique:

```python
boston_2 = boston.apply(lambda x:((x-x.mean())/(x.max(axis=0)-x.min(axis=0))).abs())
```

```python
boston_2.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston_2['crim'], bins=30)
plt.show()
```

## Normalize data using Z-score normalization technique:

```python
boston_3 = boston.apply(lambda x:((x-x.mean())/(x.std())).abs())
```

```python
boston_3.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston_3['crim'], bins=30)
plt.show()
```
